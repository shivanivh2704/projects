```python
import pandas as pd
df=pd.read_csv(r"C:\Users\dazvi\OneDrive\Desktop\cognifyz internship\Dataset .csv")
df.head()
```

```python
#find count of hotels which provides online delivery
online_delivery = df['Has Online delivery'].value_counts()
online_delivery
```

```python
#find total restaurants
total_restaurants = online_delivery.sum()
total_restaurants
```

```python
#find online_delivery_percentage
online_delivery_percentage = (online_delivery['Yes'] / total_restaurants) * 100
online_delivery_percentage
```

```python
# find data of restaurant has online delivery
df[df['Has Online delivery']=='Yes']
```

```python
# find data of restaurant doesn't has online delivery
df[df['Has Online delivery']=='No']
```

```python
# Compare the average ratings of restaurants with and without online delivery
avg_rating_online = df[df['Has Online delivery'] == 'Yes']['Aggregate rating'].mean()
avg_rating_online
```

```python
avg_rating_no_online = df[df['Has Online delivery'] == 'No']['Aggregate rating'].mean()
avg_rating_no_online
```
